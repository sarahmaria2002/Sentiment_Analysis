```python
#finding the uniqueness in every column
def unique_values(data):
    total = data.count()
    tt = pd.DataFrame(total)
    tt.columns = ['Total']
    uniques = []
    for col in data.columns:
        unique = data[col].nunique()
        uniques.append(unique)
    tt['Uniques'] = uniques
    tt['Percentage']=tt['Uniques']/tt['Total']
    return(np.transpose(tt))

unique_values(df_final)
```

```python
#finding the most frequent value
def most_frequent_values(data):
    total = data.count()
    tt = pd.DataFrame(total)
    tt.columns = ['Total']
    items = []
    vals = []
    for col in data.columns:
        itm = data[col].value_counts().index[0]
        val = data[col].value_counts().values[0]
        items.append(itm)
        vals.append(val)
    tt['Most frequent item'] = items
    tt['Frequence'] = vals
    tt['Percent from total'] = np.round(vals / total * 100, 3)
    return(np.transpose(tt))

most_frequent_values(df_final)
```

<h3>cleaning the tweets</h3>


```python
def remove_tag(string):
    text=re.sub('<.*?>','',string)
    return text
def remove_mention(text):
    line=re.sub(r'@\w+','',text)
    return line
def remove_hash(text):
     line = re.sub(r'#\w+', '', text)
     return line
def remove_newline(string):
    text=re.sub('\n','',string)
    return text
def remove_url(string):
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','',string)
    return text
def remove_number(text):
    line=re.sub(r'[0-9]+','',text)
    return line
def remove_punct(text):
    line = re.sub(r'[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~]*','',text)
    return line
def text_strip(string):
    line=re.sub('\s{2,}', ' ', string.strip())
    return line
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stopword = set(stopwords.words('english'))

def preprocess(text):
# lemmatization
    text = [lemmatizer.lemmatize(word) for word in text.split(' ')]
    text=" ".join(text)

    # stopword removal
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)

    # replace covid19 with covid
    text=text.replace('covid19','covid')
    return text

def no_spam(text):
    if 'subscri' in text:
        text=''
    return text
```

```python
india_df['refine_text']=india_df['text'].str.lower()
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_tag(str(x)))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_mention(str(x)))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_hash(str(x)))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_newline(x))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_url(x))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_number(x))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:remove_punct(x))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:text_strip(x))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:preprocess(x))
india_df['refine_text']=india_df['refine_text'].apply(lambda x:no_spam(x))

india_df['text_length']=india_df['refine_text'].str.split().map(lambda x: len(x))
```

```python
def demoji(text):

    # frequent emojis whhich will be kept
    pattern = 'üò§|üò°|üò†|üòë|üôÑ|ü§®|üò∂|üò±|üôÄ|üò≤|üòì|üò∞|üò¢|üò•|üò≠|üò™|ü§ï|üòî|üò£|üôÅ|üòí|üòñ|üòï|ü•¥|ü§í|‚òπÔ∏è|üòû|üò∑|ü§ß|üòß|üò®|üò©|ü•∫|üò¶|üòÜ|üòÄ|ü§≠|ü§©|üòå|ü•∞|üòÅ|üòò|üòÇ|üòÖ|üòä|üòù|üòô|üòá'
    for word in text:
        if re.match(pattern, word):
            continue

        # remove all other non ascii characters
        text=text.replace(word, re.sub('[^\x00-\x7f]','', word)).strip()

    return text

# apply function
india_df['refine_text']= india_df['refine_text'].apply(demoji)
```

<h2>Sentiment analysis</h2>

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
SIA = SentimentIntensityAnalyzer()

india_df["Positive"] = [SIA.polarity_scores(i)["pos"] for i in india_df["senti_text"]]
india_df["Neutral"] = [SIA.polarity_scores(j)["neu"] for j in india_df["senti_text"]]
india_df["Negative"] = [SIA.polarity_scores(k)["neg"] for k in india_df["senti_text"]]
df1 = india_df[["text", "Positive","Neutral", "Negative"]]
df1.head(3)
```

```python
pos = sum(df1["Positive"])
neu = sum(df1["Neutral"])
neg = sum(df1["Negative"])

# calculates sentiment score
def sentiment_score(pos, neu, neg):
    if (pos>neu) and (pos>neg):
        print("Positive :)")
    elif (neu>pos) and (neu>neg):
        print("Neutral :|")
    else:
        print("Negative :(")

sentiment_score(pos, neu, neg)
```

```python
# counting positive, neutral and negative tweets

sentiments_vader = []

for tweet in india_df.senti_text:
    sentiment_dict = SIA.polarity_scores(tweet)
    sentiment_dict.pop('compound', None)
    sentiments_vader.append(max(sentiment_dict , key=sentiment_dict.get))

india_df['sentiment_vader'] = sentiments_vader
india_df['sentiment_vader'].value_counts()
```

```python
#Logistic regression
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(india_df['refine_text'], india_df['sentiment_vader'], test_size=0.2, random_state=42, shuffle=False)

# Vectorizing the text data
vectorizer = TfidfVectorizer(max_features=1000)
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)

# Initializing and training the logistic regression model
logreg_model = LogisticRegression(max_iter=50)
logreg_model.fit(X_train_vect, y_train)

# Predicting on the test set
y_pred = logreg_model.predict(X_test_vect)

# Evaluating the model
logreg_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", logreg_accuracy)

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
#xgboost
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the target variable
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
# Vectorizing the text data (converting text into numerical features)
vectorizer = CountVectorizer()
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
# Converting the dataset into DMatrix format, which is optimized for XGBoost
dtrain = xgb.DMatrix(X_train_vect, label=y_train_encoded)
dtest = xgb.DMatrix(X_test_vect, label=y_test_encoded)

# Parameters for XGBoost model (you can adjust these as needed)
params = {
    'objective': 'multi:softmax',  # Multiclass classification
    'num_class': len(label_encoder.classes_),  # Number of classes
    'eta': 0.1,  # Learning rate
    'max_depth': 6,  # Maximum depth of each tree
    'eval_metric': 'merror'  # Evaluation metric
}

# Training the XGBoost model
num_rounds = 100  # Number of boosting rounds
model = xgb.train(params, dtrain, num_rounds)
# Making predictions on the testing set
y_pred_encoded = model.predict(dtest).astype(int)  # Convert float predictions to integers
# Decoding the predictions back to original labels
y_pred = label_encoder.inverse_transform(y_pred_encoded)

# Evaluating the model
xgb_accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Accuracy:", xgb_accuracy)
print("Classification Report:\n", report)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.feature_extraction.text import TfidfVectorizer

# Vectorizing the text data
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(india_df['refine_text'])

# Plot the dendrogram for a range of cluster numbers
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")
dend = shc.dendrogram(shc.linkage(X.toarray(), method='ward'), truncate_mode='level', p=10)  # Convert X to a dense array
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

# Calculate the silhouette score for different numbers of clusters
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering

best_score = -1
best_n_clusters = -1

for n_clusters in range(3, 11):
    # Create the AgglomerativeClustering object
    agg_cluster = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
    # Fit the model to the data
    cluster_labels = agg_cluster.fit_predict(X.toarray())  # Convert X to a dense array
    # Calculate the silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    # Print the silhouette score for each number of clusters
    print(f"Number of clusters: {n_clusters}, Silhouette Score: {silhouette_avg}")
    # Update the best model if a higher silhouette score is obtained
    if silhouette_avg > best_score:
        best_score = silhouette_avg
        best_n_clusters = n_clusters

# Print the best number of clusters and silhouette score
print(f"Best number of clusters: {best_n_clusters}, Best Silhouette Score: {best_score}")
```
